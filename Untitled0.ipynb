```python
import pandas as pd
df = pd.read_csv('Housing.csv')
df.head(9)
```

```python
list1 = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
df[list1] = df[list1].replace({'yes':0,'no':1})
df.tail()
```

```python
df['furnishingstatus'].unique()
```

```python
df['furnishingstatus']= df['furnishingstatus'].replace({'furnished':0,'semi-furnished':1,'unfurnished':2})
```

```python
df.tail()
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
list3 = ['price','area']
df[list3] = scalar.fit_transform(df[list3])
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df.hist(figsize=(10,10))
plt.show()
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm= LinearRegression()
lm.fit(xtrain,ytrain)
```

```python
ypredict = lm.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy = r2_score(xtest,ytest)
print(accuracy)
```
